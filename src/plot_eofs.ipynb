### Load a sample file to get the observation mask and lat, long grid

```python
gribtemplatefname = '/home/ubuntu/large-scale-climate/data/template.grb2'
aws_key = os.environ.get('AWS_KEY')
aws_secret_key = os.environ.get('AWS_SECRET_KEY')
conn = S3Connection(aws_key, aws_secret_key)
aws_key = None # So aren't saved in notebook (not sure they would be, but lets be safe)
aws_secret_key = None
with open(gribtemplatefname, 'w') as fout:
    conn.get_bucket('agittens').get_key('CSFR-O/grib2/ocnh06.gdas.1979010100.grb2').get_file(fout)

grbs = pygrib.open(gribtemplatefname)
sstoffset = 207
temps = grbs[sstoffset].values
sstlats, sstlons = grbs[sstoffset].latlons()

gribindices = range(1,41)
gribindices.append(207)
gribindices = list(reversed(gribindices))

dimsperlevel = 360*720
dims = dimsperlevel * 41
mask = np.zeros((dims,)) < 0
alllats = np.zeros((dims,))
alllons = np.zeros((dims,))

for index in range(41):
    mask[index*dimsperlevel:(index+1)*dimsperlevel] = grbs[gribindices[index]].data()[0].mask.reshape((dimsperlevel,))
    templats, templons = grbs[gribindices[index]].latlons()
    alllats[index*dimsperlevel:(index+1)*dimsperlevel] = templats.reshape((dimsperlevel,))
    alllons[index*dimsperlevel:(index+1)*dimsperlevel] = templons.reshape((dimsperlevel,))
    
SSTmask = mask[0:dimsperlevel]
```

### Code to read in the EOFs, time series, and mean temperatures from file, and plot SST fields

```python
def loadEOF(fname, numeofs, numobs):
    eofs = np.zeros((numeofs, numobs))
    with open(fname, 'r') as fin:
        fin.readline() # skip the two header lines
        fin.readline()
        for line in fin:
            (i, j, v) = line.split()
            eofs[int(j)-1, int(i)-1] = float(v)
    return eofs

def loadVec(fname, numeofs):
    vec = np.zeros((numeofs,))
    with open(fname, 'r') as fin:
        fin.readline()
        fin.readline()
        for line in fin:
            (i,j,v) = line.split()
            vec[int(j)-1] = float(v)
    return vec

def fullEOF(eof):
    eofpadded = 9999*np.ones((dims,))
    eofpadded[mask == False] = eof
    return eofpadded

def extractSST(eof):
    eofpadded = fullEOF(eof)
    return eofpadded[0:dimsperlevel]

def visualizeField(sst):
    return ma.masked_array(sst.reshape((360,720)), SSTmask.reshape((360,720)))

def plotSurfaceField(field, title):
    fig = plt.figure()
    ax = fig.add_axes([0.05,0.05,0.9,0.9])
    m = Basemap(projection='kav7',lon_0=0,resolution=None)
    m.drawmapboundary(fill_color='0.3')
    im1 = m.pcolormesh(sstlons,sstlats,visualizeField(field),shading='flat',cmap=plt.cm.jet,latlon=True)
    m.drawparallels(np.arange(-90.,99.,30.))
    m.drawmeridians(np.arange(-180.,180.,60.))
    cb = m.colorbar(im1,"bottom", size="5%", pad="2%")
    ax.set_title(title)
    fig.set_figheight(7)
    fig.set_figwidth(14)
    plt.show()
```

### Load the EOFs, the singular values, and the mean temperatures

```python
eofs = loadEOF("/home/ubuntu/large-scale-climate/data/colEOFs", 20, 6349676)
evals = loadVec("/home/ubuntu/large-scale-climate/data/evalEOFs", 20)
means = loadVec("/home/ubuntu/large-scale-climate/data/rowMeans", 6349676)
```

### Show the spectrum and the histogram of mean temperatures

```python
font = {'size'   : 16}
plt.rc('font',**font)

fig = plt.figure()
plt.plot(range(1, 21), 81/np.sum(evals*evals) * np.cumsum(evals*evals), marker="+")
plt.ylabel("Explained % of variance of centered data")
plt.xlabel("EOF index")
plt.show()
plt.hist(means[means > 0], 20)
plt.xlabel("Mean Temperature (K)")
plt.ylabel("Location counts")
plt.show()
```

### Visualize the SST EOFs and SST Mean Temperature

```python
firstssteof = extractSST(eofs[0,:])
secondssteof = extractSST(eofs[1,:])
thirdssteof = extractSST(eofs[2,:])
meansst = extractSST(means)

plotSurfaceField(firstssteof, 'SST portion of first EOF')
plotSurfaceField(secondssteof, 'SST portion of second EOF')
plotSurfaceField(thirdssteof, 'SST portion of third EOF')
plotSurfaceField(meansst, 'mean SST')
```

### Check the EOF visualization pipeline is working correctly

```python
flattemps = np.zeros((dims,))
for index in range(41):
    flattemps[index*dimsperlevel:(index+1)*dimsperlevel] = grbs[gribindices[index]].data()[0].data.reshape((dimsperlevel,))
obstemps = flattemps[mask == False]
recoveredSST = extractSST(obstemps)
plotSurfaceField(recoveredSST, 'recovered SST')
plotSurfaceField(temps.data.flatten(), 'original SST')
plotSurfaceField(recoveredSST - temps.data.flatten(), 'difference')
```
